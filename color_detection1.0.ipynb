```python
from PIL import Image
import operator
import os
import numpy as np
path1='/root/a_lilei/test2/color.txt'
path2='/root/a_lilei/data/real_picture/'
rgb_file=open(path1,'r').readlines()
rgb=[]
rgb_color=[]
r,b,w,num=0,0,0,0
#获取rgb color对应表
for i in rgb_file :
    RGB=i.split('\t')[2].split(',')
    RGB=list(map(int,RGB))
    rgb.append(RGB)
    color=i.split('\t')[-1].replace('\n','')
    rgb_color.append([RGB,color])
#循环图片文件夹，识别图片颜色    
for IMG in os.listdir(path2):
    image=Image.open(path2+IMG)
    image_color=image.getcolors(55555)
    image_colors={}
    if image_color==None:
        continue
    image_color=sorted(image_color,reverse=True)
    #计算出现次数前300的rgb值与各颜色标准rgb距离
    for j in range(0,300):
        rgb_value=abs(np.array([list(image_color[j][1])]*len(rgb_color))-np.array(rgb))
        value_sum=list(rgb_value.sum(axis=1))
        value_min=min(value_sum)         #得到最小距离
        color_index=value_sum.index(value_min)    #得到最小距离的索引
        picture_color=rgb_color[color_index][-1]
        pixel_count=image_color[j][0]
        if picture_color in image_colors:
            image_colors[picture_color]+=pixel_count
        else:
            image_colors[picture_color]=pixel_count
    #判断图片颜色
    if '红' in image_colors and image_colors['红']>1000:
        result='红'
    else:
        if '黑' in image_colors:
            image_colors['黑']-=5000
        image_colors=sorted(image_colors.items(),key=operator.itemgetter(1))
        result=image_colors[-1][0]
    if result=='黑' and len(image_colors)>2:
        if image_colors[-2][1]/image_colors[-1][1]>0.7:
            result=image_colors[-2][0]
    num+=1
    if result=='红' and result==IMG[0]:
        r+=1
    if result=='黑' and result==IMG[0]:
        b+=1
    if result=='白' and result==IMG[0]:
        w+=1
accuracy=(r+b+w)/num
print("accuracy:",accuracy)
print("red:",r)
print("white:",w)
print("black:",b)
    
        
            
        
            
    
    
        
    
            
        
            
        
        
    
    

    
    
    
    
    
    
    
```
