```python
from PIL import Image
import operator
import os
import numpy as np
name = []
image_path = "/root/a_user/car_color/photo/color/白色/"
path = '/root/a_user/car_color/color.txt'
f = open(path,'r').readlines()
colors = []
colors_rgb=[]
r= 0
b= 0
w = 0
n= 0
for i in f:
    color = i.split('\t')[1].split(' ')
    color = list(map(int,color))
    colors_rgb.append(color)   
    label = i.split('\t')[-1].replace('\n','')
    colors.append([color,label])     #记录(R,G,B) color_name对应关系
for jj in os.listdir(image_path):
    
    #print("**************************************")
    #print(jj)
    #try:
    image = Image.open(image_path+jj)

    image_color = image.getcolors(55555)
    if image_color == None:
        continue
    image_color = sorted(image_color,reverse = True)    #图像(R,G,B)count排序
    num = len(colors_rgb)   #color_name个数
    image_colors = {}
    for j in range(1,300):          #处理在image_color中排名前300的(R,G,B)
        color_differ=(np.array(colors_rgb)-np.array([list(image_color[j][1])]*num))**2 #标准R,G,B和真实R,G,B做差求平方
        differ_sum = list(color_differ.sum(axis = 1))   
        differ_min = min(differ_sum)        #记录最小差值
        color_index = differ_sum.index(differ_min)  #找到最小差值索引
        eve_color = colors[color_index][1]   #得到最小差值的颜色label
        if eve_color == '雪' or eve_color == '绿' or eve_color == '青':
            eve_color = "白"
        if '黑' in eve_color:
            eve_color = "黑"

        eve_num = image_color[j][0]
        if eve_color in image_colors:
            image_colors[eve_color] +=eve_num
        else:
            image_colors[eve_color]=eve_num

    if '红' in image_colors and image_colors['红'] > 1000:
        
        result = '红'

    else:
        if '黑' in image_colors:
            image_colors['黑'] -= 5000
        image_colors =sorted(image_colors.items(),key = operator.itemgetter(1))
        result = image_colors[-1][0]
    if result=='黑' and len(image_colors)>2:
        if image_colors[-2][1]/image_colors[-1][1] >0.7:
            result = image_colors[-2][0]


    n = n+1
    if result=='红':
        r = r+1
    if result=='黑':
        b = b+1
    if result=='白':
        w = w+1
        name.append(jj)
    #print(result)
    #print(image_colors)

   
print('红',r)
print('黑',b)
print('白',w)
print(n)
```
